**Stock Market Simulation**

Suppose you have a hedged trade (typicall NSE:NIFTY) weekly options and have risk:reward ratio of 1:4 find the probable gains over 40 weeks consistently.

**Rules**

You can skip a trade if market conditions are very bad, but if you enter you must be in the trade throughout


**Sample Size**
Experiment ran for 100,000 samples and min, max and standard deviation is provided.


```python
import random
def simulate(sample):
  outarr = []
  for i in range(sample):
    outarr.append(round(random.uniform(-1,4),2))
  return(outarr)

```

```python
exptarr = []
for expt in range(100000):
  exptarr.append(sum(simulate(40)))

```
